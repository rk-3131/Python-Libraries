```python
file = pd.read_csv('Dataset/Price for area.csv')
```

```python
file.head()
```

```python
file.sort_index(axis=0, ascending=False)
```

```python
file2 = file
```

```python
# here we have created the file2 as new view of the file view and hence any changes that we do in file2 will also reflect on 
# file too
```

```python
# to overcome this problem we have to form dataframe such that it will be copy of dataframe and not any view
```

```python
file2 = file.copy()
```

```python
file2 = file[:]
```

```python
# these two methods will make copy file2 in file database and hence it can be used to change the values which will be limited 
# to itself only
```

```python
# to access data values such that according to rowwise and columns wise indices we need to use loc function 
```
